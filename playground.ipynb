# Get text CSV from et_data_enriched

```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/kerengruteke/projects/Cognitive-State-Decoding/data/interim/et_data_enriched.csv")
```

```python
df = df.drop(columns="Unnamed: 0")
df
```

```python
df
```

```python
df[["IA_LABEL", "IA_ID", "unique_paragraph_id"]]
```

```python
 # Remove duplicates
without_duplicates = df[
    [
        "unique_paragraph_id",
        "IA_ID",
        "IA_LABEL",
    ]
].drop_duplicates()

# Group by paragraph_id, article_title, level and join all IA_LABELs (words)
text_from_et = without_duplicates.groupby(["unique_paragraph_id"])["IA_LABEL"].apply(list)

text_from_et = text_from_et.apply(lambda text: " ".join(text))
text_from_et
```

```python
texts = text_from_et.reset_index()
texts
```

```python
texts["length"] = texts["IA_LABEL"].str.len()
texts
```

```python
texts["level"] = texts["unique_paragraph_id"].str.split("_").str[2]
texts
```

```python
texts["id0"] = texts["unique_paragraph_id"].str.split("_").str[0]
texts["id1"] = texts["unique_paragraph_id"].str.split("_").str[1]
texts["id3"] = texts["unique_paragraph_id"].str.split("_").str[3]
texts["id"] = texts["id0"] + "_" + texts["id1"] + "_" +  texts["id3"]
texts
```

```python
texts.drop(columns=["id0", "id1", "id3"], inplace=True)
```

```python
texts # id = batch, article id, paragrph id
```

```python
texts.rename(columns={"IA_LABEL": "text"}, inplace=True)
```

```python
aligned = pd.read_csv("/Users/dima/Desktop/LaCC_lab/Keren's_project/aligned.csv")
aligned
```

```python
mini_df = aligned.iloc[:10].reset_index(drop=True)
mini_df
```

```python
def add_sim_cols(df):
    df["len_Ele"] = df["Text Ele Sentence"].str.len()
    df["len_Adv"] = df["Text Adv Sentence"].str.len()
    df["len_diff"] = df["len_Adv"] - df["len_Ele"]
    df["fuzz_sim"] = df[["Text Ele Sentence", "Text Adv Sentence"]].apply(
        lambda r: fuzz.ratio(r["Text Ele Sentence"], r["Text Adv Sentence"]), axis=1)
    return df

mini_df = add_sim_cols(mini_df)
mini_df
```

```python
aligned = add_sim_cols(aligned)
aligned
```
